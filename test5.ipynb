```python
import polars as pl
import matplotlib.pyplot as plt

frame = pl.read_csv("results/screening_delta_0.5.csv")
plt.plot(frame["split_rate"], frame["randomized_rate"], label="randomized")
plt.plot(frame["split_rate"], frame["splitting_rate"], label="splitting")
plt.xlabel("split rate")
plt.ylabel("screening rate")
plt.legend()
plt.show()
```
